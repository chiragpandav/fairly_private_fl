```python
import pandas as pd
import os 
from sklearn.model_selection import train_test_split

os.getcwd()
```

```python
data1="../datasets/ADULT/adult.data"
data2="../datasets/ADULT/adult.test"
```

```python
cols = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'salary'
]
df1= pd.read_csv(data1)
df2 = pd.read_csv(data2, skiprows=1, header=None) # skip first row

df1.columns = cols
df2.columns = cols
```

```python
df1.head()
```

```python
df2.head()
```

```python
merged_df =pd.concat([df1, df2], axis=0)

merged_df['workclass'] = merged_df['workclass'].str.strip()
merged_df['education'] = merged_df['education'].str.strip()
merged_df['marital-status'] = merged_df['marital-status'].str.strip()
merged_df['occupation'] = merged_df['occupation'].str.strip()
merged_df['relationship'] = merged_df['relationship'].str.strip()

merged_df['race'] = merged_df['race'].str.strip()
merged_df['sex'] = merged_df['sex'].str.strip()
merged_df['native-country'] = merged_df['native-country'].str.strip()
merged_df.tail()
```

```python
merged_df['salary'] = merged_df['salary'].str.replace('<=50K', '<=50K').str.replace('>50K', '>50K').str.replace(' ', '').str.replace('.', '')
```

```python
#Removing row with '?' 
df_cleaned = merged_df[~merged_df.apply(lambda x: x.astype(str).str.contains('\?')).any(axis=1)]

# View the cleaned DataFrame
df_cleaned
```

```python
merged_df = df_cleaned.copy()
```

```python
merged_df.isnull().sum()
```

```python
merged_df
```

```python
pd.unique(merged_df["occupation"])
```

```python
merged_df.dropna(inplace=True)
len(merged_df)
```

```python
merged_df['race_sex'] = merged_df['race'] + '_' + merged_df['sex']

# Step 2: Stratified Sampling for df1 (10,000 data points)
df1_sample, remaining_data = train_test_split(
    merged_df, test_size=(1 - 6000 / len(merged_df)), stratify=merged_df['race_sex'], random_state=42
)
df1_sample = df1_sample.drop(columns=['race_sex'])
df1_sample = df1_sample.reset_index(drop=True)
df1_sample
```

```python
len(remaining_data)
```

```python
df2_sample, remaining_data = train_test_split(
    remaining_data, test_size=(1 - 10000 / len(remaining_data)), stratify=remaining_data['race_sex'], random_state=42
)
df2_sample = df2_sample.drop(columns=['race_sex'])

df2_sample = df2_sample.reset_index(drop=True)

df2_sample
```

```python
len(df2_sample),len(df1_sample)
```

```python
len(remaining_data)
```

```python
dfs_2000 = []

while len(remaining_data) >= 2000:
    sample_2000, remaining_data = train_test_split(
        remaining_data, test_size=(1 - 2000 / len(remaining_data)), stratify=remaining_data['race_sex'], random_state=42
    )
    dfs_2000.append(sample_2000)

# Handle remaining data points (if any) as the last DataFrame
if len(remaining_data) > 0:
    dfs_2000.append(remaining_data)

dfs_2000 = [df.drop(columns=['race_sex']) for df in dfs_2000]
dfs_2000 = [df.reset_index(drop=True) for df in dfs_2000]
```
